```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import sys

data = np.loadtxt('ex2data1.txt', delimiter = ',')
```

```python
X = np.c_[np.ones([data.shape[0],1]),data[:,0:2]];
y = data[:,2].reshape([X.shape[0],1])
```

```python
def mapp(cc):
    if(cc==0.): return 'g'
    if(cc==1.): return 'r'
coll=[mapp(cc) for cc in y[:,0]]
```

```python
col = ['red','green']
#plt.scatter(X[:,0],X[:,1],c = coll)
plt.scatter(X[:,1],X[:,2],c = list(y[:,0]),cmap=matplotlib.colors.ListedColormap(col));
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def computeCost(X,y,theta=np.zeros([X.shape[1],1])):
    m = y.size;
    h = sigmoid(X.dot(theta))
    cost = 1/m*(-y.T.dot(np.log(h))- (1-y).T.dot(np.log(1-h)))    
    return cost

computeCost(X,y,theta = [[-23.3],[0.],[0.201472]])
```

```python
def gradientDescent(X,y,theta = np.zeros([X.shape[1],1]),alpha = 0.001, num_iter = 100):
    cost = np.zeros([num_iter,1])
    m = y.size;

    for _ in range(num_iter):
        h = sigmoid(X.dot(theta))

        gradient = 1/m*(X.T.dot(h-y))
        theta = theta - alpha*gradient

        cost[_] = computeCost(X,y,theta)
    return theta, cost
```

```python
theta, cost = gradientDescent(X,y, theta = [[-24.3],[0.5],[0.201472]] )
print(computeCost(X,y,theta))
plt.plot(cost); plt.ylabel('cost');
print('theta:', theta.ravel())
```

```python
xx = range(int(min(X[:,1])), int(max(X[:,1])))
yy = (-theta[0]-theta[1]*xx)/theta[2]
plt.plot(xx,yy)
```
