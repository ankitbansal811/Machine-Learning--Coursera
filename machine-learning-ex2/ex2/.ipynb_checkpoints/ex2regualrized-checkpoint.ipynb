```python
# %load ../../standard_import.txt
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from scipy.optimize import minimize

from sklearn.preprocessing import PolynomialFeatures
```

```python
data = np.loadtxt('ex2data2.txt', delimiter = ',')
X = data[:,0:2]
y = np.c_[data[:,2]]
```

```python
col = ['red','green']
#plt.scatter(X[:,0],X[:,1],c = coll)
plt.scatter(X[:,0],X[:,1],c = list(y[:,0]),cmap=mpl.colors.ListedColormap(col));
```

```python
# Note that this function inserts a column with 'ones' in the design matrix for the intercept.
poly = PolynomialFeatures(6)
XX = poly.fit_transform(data[:,0:2])
XX.shape
```

```python
def sigmoid(z):
    return(1 / (1 + np.exp(-z)))
```

```python
def costFunction(theta, reg, X, y):
    m = y.size
    h = sigmoid(X.dot(theta))
    
    J = -1*(1/m)*(np.log(h).T.dot(y)+np.log(1-h).T.dot(1-y))
    J = J + reg/2/m*np.sum(np.square(theta[1:]))
               
    if np.isnan(J[0]):
        return(np.inf)
    return(J[0])
```

```python
def gradient(theta, reg, X, y):
    m = y.size
    h = sigmoid(X.dot(theta.reshape(-1,1)))
    
    grad =(1/m)*X.T.dot(h-y) + reg/m*np.r_[[[0]],theta[1:].reshape(-1,1)]

    return(grad.flatten())
```

```python
initial_theta = np.zeros(XX.shape[1])
cost = costFunction(initial_theta,1, XX, y)
grad = gradient(initial_theta,1, XX, y)
print('Cost: \n', cost)
print('Grad: \n', grad)
```

```python
def predict(theta, X, threshold=0.5):
    p = sigmoid(X.dot(theta.T)) >= threshold
    return(p.astype('int'))
```

```python
fig, axes = plt.subplots(1,3, sharey = True, figsize=(17,5))

# Decision boundaries
# Lambda = 0 : No regularization --> too flexible, overfitting the training data
# Lambda = 1 : Looks about right
# Lambda = 100 : Too much regularization --> high bias

# Plot decisionboundary
x1_min, x1_max = X[:,0].min(), X[:,0].max(),
x2_min, x2_max = X[:,1].min(), X[:,1].max(),
xx1, xx2 = np.meshgrid(np.linspace(x1_min, x1_max), np.linspace(x2_min, x2_max))

for i, C in enumerate([0, 1, 100]):
    # Optimize costFunctionReg
    res2 = minimize(costFunction, initial_theta, args=(C, XX, y), method=None, jac=gradient, options={'maxiter':3000})
    
    # Accuracy
    accuracy = 100*sum(predict(res2.x, XX) == y.ravel())/y.size    

    # Scatter plot of X,y
    #plotData(data2, 'Microchip Test 1', 'Microchip Test 2', 'y = 1', 'y = 0', axes.flatten()[i])
    axes.flatten()[i].scatter(X[:,0],X[:,1],c = list(y[:,0]),cmap=mpl.colors.ListedColormap(col));
    

    h = sigmoid(poly.fit_transform(np.c_[xx1.ravel(), xx2.ravel()]).dot(res2.x))
    h = h.reshape(xx1.shape)
    axes.flatten()[i].contour(xx1, xx2, h, [0.5],linewidths=1, colors='g');       
    axes.flatten()[i].set_title('Train accuracy {}% with Lambda = {}'.format(np.round(accuracy, decimals=2), C))
```
