```python
import numpy as np
import pandas as pd
import scipy.io
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from sklearn.linear_model import LinearRegression as LR
from sklearn.preprocessing import PolynomialFeatures

mat = scipy.io.loadmat('ex5data1.mat')
X = mat['X'];  y = mat['y']
Xval = mat['Xval'];  yval = mat['yval']
Xtest = mat['Xtest'];  ytest = mat['ytest']
```

```python
plt.scatter(X,y,c = 'r',marker = 'x')
plt.xlabel('Change in water level (x)');plt.ylabel('Water flowing out of the dam (y)');
```

```python
def Cost(theta,X,y,reg):
    theta = theta.reshape([-1,1])
    m = y.size;
    X = np.c_[np.ones([m,1]), X]
    J = 1/(2*m)*np.sum((y-X.dot(theta))**2) + reg/2/m*(np.sum(theta[1:]**2));   
    return J

def Gradient(theta,X,y,reg):
    theta = theta.reshape([-1,1])
    m = y.size;
    X = np.c_[np.ones([m,1]), X]
    grad = -1/m*X.T.dot(y-X.dot(theta).reshape([-1,1])) + reg/m*np.r_[[[0]],theta[1:]]
    return grad.flatten()
```

```python
theta = np.array([[1],[1]])
```

```python
regr = LR()
regr.fit(X.reshape(-1,1),y);
```

```python
reg = 1
result = minimize(Cost, theta,args = (X,y,reg), method = None, jac = Gradient, options = {'maxiter':10})
xx = np.linspace(min(X),max(X),50).reshape([-1,1])
plt.plot(xx, regr.intercept_+xx*regr.coef_,label = 'LR')
plt.scatter(X,y,c = 'r',marker = 'x',label = 'Train data')
theta_LR = np.array([regr.intercept_, regr.coef_])
theta_min = result.x
plt.plot(xx, np.c_[np.ones([xx.shape[0],1]),xx].dot(theta_min),label = 'Minimize')
plt.legend(loc=4);
cost_LR = Cost(theta_LR,X,y,0)
cost_min = Cost(theta_min,X,y,0)
plt.text(min(X), max(y), 'Cost for LR:{:0.2f} \nCost for minimize:{:0.2f} '.format(cost_LR,cost_min), horizontalalignment='left', verticalalignment='top',fontsize = 15);
```

```python
def LearningCurve(Xtrain,ytrain,Xval,yval,reg):
    m = ytrain.size
    error_train = np.zeros([m,1]);
    error_val = np.zeros([m,1]);
    
    for i in range(m):
        X = Xtrain[:i+1]; y = ytrain[:i+1]
        theta = np.random.rand(X.shape[1]+1,1)
        result = minimize(Cost, theta,args = (X,y,reg), method = None, jac = Gradient, options = {'maxiter':10})
        theta = result.x
        error_train[i]= Cost(theta,X,y,0)
        error_val[i]= Cost(theta,Xval,yval,0)
        
    return error_train, error_val
```

```python
error_train, error_val = LearningCurve(X,y,Xval,yval,reg)
plt.plot(range(y.size),error_train,range(y.size),error_val);
plt.ylim((0,150));   plt.xlim((0,y.size));
```

```python
poly = PolynomialFeatures(8)
X_poly = poly.fit_transform(X)[:,1:]
```

```python
def featureNormalize(X_poly):    
    mu = np.mean(X_poly,0); #mu[0] = 0;
    std = np.std(X_poly,0); #std[0] = 1;
    X_poly = (X_poly - mu)/std    
    return X_poly, mu, std

[X_poly, mu, std] = featureNormalize(X_poly)
Xval_poly = (poly.fit_transform(Xval)[:,1:] - mu)/std
Xtest_poly = (poly.fit_transform(Xval)[:,1:] - mu)/std
```

```python
theta = np.random.rand(X_poly.shape[1]+1,1)

result = minimize(Cost, theta,args = (X_poly,y,reg), method = None, jac = Gradient, options = {'maxiter':10})
xx = np.linspace(min(X)-20,max(X)+20,50).reshape([-1,1])
xx_poly = (poly.fit_transform(xx)[:,1:] - mu)/std
plt.scatter(X,y,c = 'r',marker = 'x',label = 'Train data')

theta_min = result.x
plt.plot(xx, np.c_[np.ones([xx_poly.shape[0],1]),xx_poly].dot(theta_min),label = 'Minimize')
plt.legend(loc=4);
cost_min = Cost(theta_min,X_poly,y,0)
plt.text(min(X), max(y), 'Cost for minimize:{:0.2f} '.format(cost_min), horizontalalignment='left', verticalalignment='center',fontsize = 15);
```

```python
error_train, error_val = LearningCurve(X_poly,y,Xval_poly,yval,reg)
plt.plot(range(y.size),error_train,range(y.size),error_val);
plt.ylim((0,100));   plt.xlim((0,y.size));
```

```python
def ValidationCurve(Xtrain,ytrain,Xval,yval,reg):
    m = reg.size
    error_train = np.zeros([m,1]);
    error_val = np.zeros([m,1]);
    
    for i in range(m):
        theta = np.random.rand(Xtrain.shape[1]+1,1)
        result = minimize(Cost, theta,args = (Xtrain,ytrain,reg[i]), method = None, jac = Gradient, options = {'maxiter':10})
        theta = result.x
        error_train[i]= Cost(theta,Xtrain,ytrain,0)
        error_val[i]= Cost(theta,Xval,yval,0)
        
    return error_train, error_val
```

```python
reg = np.array([0, 0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3, 10])
error_train, error_val = ValidationCurve(X_poly,y,Xval_poly,yval,reg)
plt.plot(reg,error_train,reg,error_val);
plt.xlim((0,reg.size));
```
